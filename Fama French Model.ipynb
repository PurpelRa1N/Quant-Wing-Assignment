```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import statsmodels.api as sm 
import pandas_datareader as web
```

```python
start = '2015-09-01'
end = '2021-01-01'
```

```python
fundsret = web.get_data_yahoo('^NSEI', start, end)['Adj Close'].pct_change(1)
fundsret = pd.DataFrame(fundsret)
```

```python
fundsret.head()
```

```python
fundsret_monthly = fundsret.resample('M').agg(lambda x: (x+1).prod()-1)
fundsret_monthly = fundsret_monthly.rename(columns={"Adj Close" : "Monthly Returns in %"})
```

```python
fundsret_monthly
```

```python
factors = pd.read_csv('2021-03_Size_Value_Portfolio_Returns_Monthly_SurvivorshipBiasAdjusted.csv')
factors = pd.DataFrame(factors)
factors.set_index("Date", inplace=True)
factors = factors
factors = factors.iloc[263:327]
factors
```

```python
factors['SMB in %'] = (factors['Portfolios SV returns (%)'] + factors['Portfolios SN returns (%)'])
factors["HML in %"] = (factors['Portfolios SV returns (%)'] + factors['Portfolios BV returns (%)'])
factors
```

```python
fundsret_monthly.index = factors.index
merge = pd.merge(fundsret_monthly, factors, on='Date')
merge
```

```python
risk = pd.read_csv('2021-03_FourFactors_and_Market_Returns_Monthly_SurvivorshipBiasAdjusted.csv')
risk = pd.DataFrame(risk[['MF', 'RF']])
risk = risk.iloc[263:327]
risk.index = merge.index
merge = pd.merge(merge, risk, on='Date')
```

```python
merge['Returns - RF'] = merge['Monthly Returns in %'] - merge['RF']
merge
```

```python
y = merge['Returns - RF']
X = merge[['MF', 'SMB in %', 'HML in %']]
x_sm = sm.add_constant(X)
```

```python
model = sm.OLS(y, x_sm)
results = model.fit()
results.summary()
```
