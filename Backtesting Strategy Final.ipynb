```python
import yfinance as yf
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = yf.download('RELIANCE.NS', start='2015-01-01', end='2022-01-01')
```

```python
data['MA100'] = data['Adj Close'].rolling(100).mean()
data
```

```python
data = data.dropna()
data
```

```python
plt.figure(figsize=(12,8))
plt.plot(data['Adj Close'])
plt.plot(data['MA200'])
```

```python
pd.options.mode.chained_assignment = None
```

```python
data['Price Change'] = data['Adj Close'].pct_change()
```

```python
data = data.dropna()
data
```

```python
data['Upmove'] = data['Price Change'].apply(lambda x:x if x>0 else 0)
```

```python
data['Downmove'] = data['Price Change'].apply(lambda x:x if x<0 else 0)
```

```python
data['Avg Up'] = data['Upmove'].ewm(span=19).mean()
data['Avg Down'] = data['Downmove'].ewm(span=19).mean()
data = data.dropna()
data
```

```python
data['RS'] = data['Avg Up']/data['Avg Down']
```

```python
data['RSI'] = data['RS'].apply(lambda x: 100-(100/(x+1)))
```

```python
data.loc[(data['Adj Close'] > data['MA200']) & (data['RSI'] < 30), 'Buy'] = 'Yes'
```

```python
data.loc[(data['Adj Close'] < data['MA200']) | (data['RSI'] > 30), 'Buy'] = 'No'
```

```python
data
```

```python
PnL = []


for i in range(len(data) - 12):
    if "Yes" in data['Buy'].iloc[i]:
        for j in range(1,11):
            if data['RSI'].iloc[i+j] > 40:
                PnL.append(data['Open'].iloc[i+j+1]- data['Open'].iloc[i+1])
                break
            
            if data['RSI'].iloc[i+j] < 40:
                PnL.append(data['Open'].iloc[i+12] - data['Open'].iloc[i+1])
                break
```
