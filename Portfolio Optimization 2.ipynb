```python
data = data[1:]
data
```

```python
correlation = data.corr()
correlation = np.abs(correlation)
correlation["Stock 1"] = correlation.index
correlation = correlation.melt(id_vars = "Stock 1", var_name = "Stock 2", value_name = "Abs Corr")
correlation = correlation[correlation["Stock 1"] < correlation["Stock 2"]].dropna()
correlation = correlation.sort_values("Abs Corr", ascending = True)
correlation
```
